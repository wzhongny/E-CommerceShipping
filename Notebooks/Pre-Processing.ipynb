```python
#imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#load data
df = pd.read_csv('df.csv')
df.head()
```

```python
df.dtypes
```

All of our data is numerical so we can just do a train test split on this. We will do two train test splits, one with On_time as the label and one with Customer_rating as the label.

```python
# train test split for on_time model
X_train_ot, X_test_ot, y_train_ot, y_test_ot = train_test_split(df.drop(columns='On_time'), df.On_time, test_size=0.2, random_state=21)
```

```python
X_train_ot.shape, X_test_ot.shape
```

```python
y_train_ot.shape, y_train_ot.shape
```

```python
# scale train set
scaler = StandardScaler()
scaler.fit(X_train_ot)
X_train_ot_scaled = scaler.transform(X_train_ot)
X_train_ot_scaled = pd.DataFrame(X_train_ot_scaled, columns=X_train_ot.columns)
X_train_ot_scaled.head()
```

```python
# scale test set
X_test_ot_scaled = scaler.transform(X_test_ot)
X_test_ot_scaled = pd.DataFrame(X_test_ot_scaled, columns=X_test_ot.columns)
X_test_ot_scaled.head()
```

```python
# do train test split and scale data now for customer_rating model
X_train_cr, X_test_cr, y_train_cr, y_test_cr = train_test_split(df.drop(columns='Customer_rating'), df.Customer_rating, test_size=0.2, random_state=21)
scaler = StandardScaler()
scaler.fit(X_train_cr)
X_train_cr_scaled = scaler.transform(X_train_cr)
X_train_cr_scaled = pd.DataFrame(X_train_cr_scaled, columns=X_train_cr.columns)
X_test_cr_scaled = scaler.transform(X_test_cr)
X_test_cr_scaled = pd.DataFrame(X_test_cr_scaled, columns=X_test_cr.columns)
```

```python
X_train_cr_scaled.head()
```

```python
X_test_cr_scaled.head()
```

```python
# save data sets
X_train_ot.to_csv('X_train_ot.csv')
X_test_ot.to_csv('X_test_ot.csv')
y_train_ot.to_csv('y_train_ot.csv')
y_test_ot.to_csv('y_test_ot.csv')
X_train_ot_scaled.to_csv('X_train_ot_scaled.csv')
X_test_ot_scaled.to_csv('X_test_ot_scaled.csv')
X_train_cr.to_csv('X_train_cr.csv')
X_test_cr.to_csv('X_test_cr.csv')
y_train_cr.to_csv('y_train_cr.csv')
y_test_cr.to_csv('y_test_cr.csv')
X_train_cr_scaled.to_csv('X_train_cr_scaled.csv')
X_test_cr_scaled.to_csv('X_test_cr_scaled.csv')
```
